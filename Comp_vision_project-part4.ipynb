```python
from IPython.display import display, Image
import cv2
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow

#read bounding boxes from file and convert to mediapipe format
def read_and_convert_bounding_boxes(file_path, image_width, image_height):
    pixel_coordinates = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.split()
            class_id, x_center_norm, y_center_norm, width_norm, height_norm = map(float, parts[:5])
            x_center, y_center = x_center_norm * image_width, y_center_norm * image_height
            width, height = width_norm * image_width, height_norm * image_height
            x1 = int((x_center - width / 2))
            y1 = int((y_center - height / 2))
            w = int(width)
            h = int(height)
            pixel_coordinates.append((x1, y1, w, h))
    return pixel_coordinates


#image
image_path = '/content/yolov7/runs/detect/exp/youtube-71_jpg.rf.8c3ab1b5d23cd48cd392608ad8ef3a27.jpg'
image = cv2.imread(image_path)
cv2_imshow(image)
txt_file_path = '/content/yolov7/runs/detect/exp/labels/youtube-71_jpg.rf.8c3ab1b5d23cd48cd392608ad8ef3a27.txt'
image_width, image_height = 640, 640

pixel_coordinates = read_and_convert_bounding_boxes(txt_file_path, image_width, image_height)





```

```python
     # Convert to grayscale and apply edge detection - before I saw that I can extract label coordinates
from google.colab.patches import cv2_imshow

def read_bounding_boxes(txt_file):
    bounding_boxes = []
    with open(txt_file, 'r') as f:
        for line in f.readlines():
            if len(line) > 1:  # Skip empty lines
                parts = line.split()
                x_center, y_center, width, height = map(float, parts[1:5])
                x1 = x_center - width / 2
                y1 = y_center - height / 2
                bounding_boxes.append((x1, y1, width, height))
    return bounding_boxes

image = cv2.imread('/content/yolov7/runs/detect/exp3/youtube-4_jpg.rf.3241e3f563d88c7fc39f569beb08ae82.jpg')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray_image, threshold1=100, threshold2=200)

cv2_imshow(gray_image)
#contours
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2_imshow(edges)


#filter contours
def filter_contours(contours, min_area=1000, max_area=50000, min_aspect_ratio=0.2, max_aspect_ratio=1.0):
    valid_contours = []
    for contour in contours:
        area = cv2.contourArea(contour)
        x, y, w, h = cv2.boundingRect(contour)
        aspect_ratio = float(w) / h
        if min_area < area < max_area and min_aspect_ratio < aspect_ratio < max_aspect_ratio:
            valid_contours.append(contour)
    return valid_contours

filtered_contours = filter_contours(contours)

```

```python
# Initialize MediaPipe Pose - images
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.1)
mp_drawing = mp.solutions.drawing_utils

#Pose estimation on a cropped image
def apply_pose_estimation_to_crop(image, box):
    x, y, w, h = box
    crop_img = image[y:y+h, x:x+w]
    cv2_imshow(crop_img)
    results = pose.process(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB))
    if results.pose_landmarks:
        mp_drawing.draw_landmarks(crop_img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
    return crop_img

for contour in filtered_contours:
    x, y, w, h = cv2.boundingRect(contour)
    crop_img = apply_pose_estimation_to_crop(image, (x, y, w, h))
    cv2_imshow(crop_img)  #each cropped image with pose estimation


for i, contour in enumerate(filtered_contours):
    x, y, w, h = cv2.boundingRect(contour)
    crop_img = apply_pose_estimation_to_crop(image, (x, y, w, h))
    cv2_imshow(crop_img)  #each cropped image with pose estimation
```
